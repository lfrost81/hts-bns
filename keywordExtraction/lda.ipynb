```python
from gensim import corpora, models
import gensim
from beans import Beans, Bean
```

```python
field = [
    ("DOCID", "docid", "text", "primary key"), \
    ("DATE", "date", "int"), \
    ("URL", "url", "text"), \
    ("CATEGORY", "category", "text"), \
    ("TITLE", "title", "text"), \
    ("SNIPPET", "snippet", "text"), \
    ("CONTENT", "content", "text"), \
    ("SEQ", "token", "text") \
    ]
```

```python
print("=" * 10 + "topic analysis start" + "=" * 10)
```

```python
bean = Beans()
```

```python
bean.init("./data/beans_komoran", field, False)
```

```python
bean.read_file()
```

```python
result = bean.executeQuery("select token from document")
```

```python
documents = [ list(c) for c in zip(*result) ][0]
```

```python
print (documents[0])
```

```python
texts = []
for doc in documents :
    texts.append( doc.split() )
```

```python
# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary( texts )
```

```python
# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
topic_count = 30
```

```python
# generate LDA model
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=topic_count, id2word=dictionary, passes=20,iterations=100)
```

```python
print ( ldamodel.get_document_topics(dictionary.doc2bow(texts[0]), minimum_probability=0.5) )
```

```python
print("=" * 10 + "topic analysis end" + "=" * 10)
```

```python
ldamodel.print_topics()
```

```python
id2name = dict()

with open("./data/id2name") as f:
    for line in f:
        item = line.strip().split('|')
        id2name[int(item[0])] = item[1]
```

```python
keyword2name = dict()

with open("./data/merchant2.txt") as ifs_m2k:
    for line in ifs_m2k:
        item = line.strip().split('\t')
        keyword2name[item[1].strip()] = item[0].strip()
```

```python
def getBaseMerchantName(keyword):
    return keyword2name[keyword.strip()]
```

```python
def getMerchantName(mid):
    return getBaseMerchantName(id2name[mid])
```

```python
merchantListByTopic = [None]*topic_count

for i in range( 0, topic_count ) :
    merchantListByTopic[i] = set()
```

```python
for c, text in enumerate( texts ) :
    for topic in ldamodel.get_document_topics( dictionary.doc2bow( text ), minimum_probability=0.3 ) :
        try :
            merchantListByTopic[ topic[0] ].add( getMerchantName(c) )
        except :
            continue
        #print( topic[0], getMerchantName(c)  )
```

```python
labelByTopic = [ "합리주의","식도락가","스타일리쉬","대중적인","사회적인","얼리어댑터","커피홀릭","연예인병","웰빙족","패셔너블한" ]

for i in range ( 0 , topic_count ) :
    print( "%d = [%s]\n{%s}\n" % (i, ", ".join( merchantListByTopic[i] ), ldamodel.print_topic(i) ) )
```
